Критерий


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ncf, norm, f

alpha = 0.05
n, m = 139, 1000
Sx, Sy = 5.722, 6.161
Sx_w, Sy_w = 4.612, 5.055

delta1 = f.ppf(alpha / 2, n-1, m-1)
delta2 = f.ppf(1 - alpha / 2, n-1, m-1)

print(f"Критич. значения: {delta1} - {delta2}")
```

=> нент оснований отвергнуть гипотезу H₀
----
теперь зависимость мощности от отношения дисперсий


```python
a = np.linspace(0.5, 2, 100)
W = 1 - f.cdf(delta2, n-1, m-1, scale=a) + f.cdf(delta1, n-1, m-1, scale=a)

plt.figure(figsize=(10, 6))
plt.plot(a, W, label="Мощность критерия", color="orange")
plt.axvline(x=1, color="green", linestyle="-", label="H₀: дисперсии равны")
plt.xlabel("Отношение дисперсий (фи / пси)")
plt.ylabel("Мощность критерия")
plt.legend()
plt.grid()
plt.show()
```
