```python
# ИНТЕРВАЛ ЧЕРЕЗ ТОЧНИ МЕТОД
```

```python
a = 0.05

ЛЕОВ = np.max(X) / (1 + (1 - a / 2) ** (1 / N))
ПРАВ = np.max(X) / (1 + (a / 2) ** (1 / N))
print()
print("Интервал через ТОЧНЫЙ метод: ", "(", round(ЛЕОВ, 4), ',', round(ПРАВ, 4), ')')
print("Размах = ", round(ПРАВ - ЛЕОВ, 4), "\n")
```
